## 7. Concatenate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
import numpy as np
a = np.array([1, 2, 3]);
b = np.array([4, 5, 6]);
concat=np.concatenate((a,b))
print(concat)
```

## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import pandas as pd
dates=pd.date_range(start='1-1-2023',end='02-10-2023')
for val in dates:
    print(val)
```
